```python
#!pip install statsmodels
import pandas as pd
from statsmodels.tsa.seasonal import seasonal_decompose


data_orig = pd.read_csv('apparelCPI2021.csv')

data_orig['Date'] = pd.to_datetime(data_orig['Date']) # convert date column to DateTime
#ax = data_orig.plot(x='Date', y='ApparelCPI', figsize=(12,6))

data_orig.set_index('Date', inplace=True)

analysis = data_orig[['ApparelCPI']].copy()


decompose_result_mult = seasonal_decompose(analysis, model="multiplicative")

trend = decompose_result_mult.trend
seasonal = decompose_result_mult.seasonal
residual = decompose_result_mult.resid

#decompose_result_mult.plot();

seasonal=seasonal*100
seasonal.tail(12).plot()
seasonal.to_csv("SeasonalityIndex2021.csv")
```
